```python
import random as rnd 
import matplotlib.pyplot as plt
import numpy as np
n = 0
innelista_x = []
innelista_y = []
utelista_x = []
utelista_y = []
for k in range(1000):
    x = rnd.uniform(-1,1) 
    y = rnd.uniform(-1,1)
    #print("(x,y) = (",x,",", y,")" ) 

    if x**2+y**2 <= 1: 
        n += 1 
        innelista_x.append(x) 
        innelista_y.append(y) 
    else:
        utelista_x.append(x)
        utelista_y.append(y)

andel= (n/(k+1))*4 
print(andel)

plt.plot(innelista_x,innelista_y,".")
plt.plot(utelista_x,utelista_y,".r")
plt.show()
```
